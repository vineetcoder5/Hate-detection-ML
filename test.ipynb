```python
import pandas as pd
import numpy as np
from google.colab import drive
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from joblib import dump, load
from sklearn.metrics import precision_score, f1_score, recall_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
```

```python
# acessing the file from google drive
file_path = "/content/drive/MyDrive/hate_speech_dataset/train.csv"
main_df = pd.read_csv(file_path)
```

```python
# making a copy of data set so that it will not hamper orginal data
df = main_df.copy()
df["hate"] = -1
```

```python
# this is the condition to check to update hate coloumb value
condition = (df['toxic'] == 1) | (df['severe_toxic'] == 1) | (df['obscene'] == 1) | (df['threat'] == 1) | (df['insult'] == 1) | (df['identity_hate'] == 1)

# Update the 'hate' column based on the condition
df['hate'] = np.where(condition, 1, 0)
df.head(10)
```

```python

# Select target
y = df.hate
X_text = df['comment_text']

count_vectorizer = CountVectorizer()
count_vectorizer.fit(X_text)
```

```python
vector = count_vectorizer.transform(X_text)
```

```python
# Divide data into training and validation subsets
X_train, X_valid, y_train, y_valid = train_test_split(vector, y, train_size=0.7, test_size=0.3, random_state=1)

# Model training and evaluation
# model = RandomForestClassifier(n_estimators=100, random_state=1)
# model = DecisionTreeClassifier(random_state=1)
# model = KNeighborsClassifier(n_neighbors=100)
# model = LogisticRegression(random_state=1)
base_classifier = DecisionTreeClassifier(random_state=2)
model = AdaBoostClassifier(base_classifier, random_state=0)
model.fit(X_train, y_train)
preds = model.predict(X_valid)

# Evaluate the model
accuracy = accuracy_score(y_valid, preds)
print(f"Accuracy: {accuracy}")
```

```python

l = ["i love you","i hate you"]

# Fit and transform the testing data
X_counts = count_vectorizer.transform(l)

# Make predictions using the trained model
preds = model.predict(X_counts)

# Display the predictions
print("Predictions:", preds)
# dump(model, 'countvector_1.pkl')
```

```python
# preds = model.predict(X_valid)
precision = precision_score(y_valid, preds)
print(f'Precision: {precision}')

# Calculate F1 score
f1 = f1_score(y_valid, preds)
print(f'F1 Score: {f1}')

# Calculate recall
recall = recall_score(y_valid, preds)
print(f'Recall: {recall}')
```
